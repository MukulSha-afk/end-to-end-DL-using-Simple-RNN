```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
##define the vocalbulary size
voc_sie = 10000
```

```python
one_hot_repr=[one_hot(word,voc_sie)for word in sent]
one_hot_repr
```

```python
##word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, maxlen=sent_length, padding='pre')
embedded_docs
```

```python
## feature representtation
dim = 10
model=Sequential()
model.add(Embedding(voc_sie,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
